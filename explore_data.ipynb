# Explorative Datenanalyse

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
# Erste “Insights”, d.h. explorative Datenanalayse inkl. Charts (Verteilungen, erste Korrelationen, Nullwerte, etc.,)
current_dir = os.getcwd().replace("\\", "/")
df = pd.read_csv(current_dir + "/data/train.csv")
df
```

```python
#exploratory data analysis
df.info()
df.describe()
```

```python
#plot heatmap for missing values
sns.heatmap(df.isnull(), cbar=False)
```

```python
#histogram for all columns
df.hist(figsize=(20, 20))
plt.show()
```

```python
#correlation matrix
correlation=df.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Group by race_group and calculate EFS percentage
grouped = df.groupby('race_group').agg(
    total=('efs', 'count'),
    efs_count=('efs', 'sum')
)
grouped['efs_percentage'] = (grouped['efs_count'] / grouped['total']) * 100

# Reset index for plotting
grouped = grouped.reset_index()

# Plot the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='race_group', y='efs_percentage', data=grouped, palette='viridis')
plt.title('Percentage of Event-Free Survival (EFS) by Race Group')
plt.xlabel('Race Group')
#angle the x-axis labels
plt.xticks(rotation=45)
plt.ylabel('EFS Percentage')
plt.ylim(0, 100)  # Ensure percentages are within 0-100
plt.show()
```

```python
#plot the distribution of the target variable "age_at_hct"
plt.figure(figsize=(8, 6))
sns.histplot(df['age_at_hct'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age at HCT')
plt.xlabel('Age at HCT')
plt.ylabel('Frequency')
plt.show()
```

```python
#bar plot to show the columns with missing values
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values.sort_values(inplace=True)
plt.figure(figsize=(8, 6))
missing_values.plot(kind='barh', color='skyblue')
plt.title('Columns with Missing Values')
plt.xlabel('Columns')
plt.ylabel('Number of Missing Values')
plt.show()
```

```python
# Eine neue Spalte erstellen, um NaN-Werte zu markieren
df['age_group'] = df['age_at_hct'].apply(lambda x: 'Missing' if pd.isnull(x) else 'Not Missing')

# Plot: Balkendiagramm der fehlenden und nicht fehlenden Werte
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df, 
    x='age_at_hct', 
    hue='age_group', 
    bins=20, 
    kde=False, 
    palette='viridis',
    multiple="stack"
)

# Titel und Achsenbeschriftungen
plt.title('Distribution of Age at HCT with Missing Values Highlighted')
plt.xlabel('Age at HCT')
plt.ylabel('Frequency')
plt.legend(title='Age Group')
plt.show()
```
